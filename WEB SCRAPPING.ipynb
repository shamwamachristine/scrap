## WEBSITE

## REQUEST

```python
response = requests.get(website)
```

## SOUP OBJECT

```python
soup = BeautifulSoup(response.content,'html.parser')
```

## RESULTS

```python
results = soup.find_all('div', class_="product-item d-flex col-6 col-sm-4 col-lg-3")
```

## all columns

```python
art_name = []

for result in results:
    art_name.append(result.find('h5', {'class': 'card-title product-card-name'}).get_text())
    
```

```python
len(art_name)
```

```python
price = []


for result in results:
    price.append(result.find('span',{'class' : 'stockrecord-price-current'}).get_text().strip())
```

```python
len(price)
```

```python
art_supplies = {'Art Supplies':art_name,'Art Price':price}
```

## pagination: Scrapping multiple pages

```python
for i in range(1,42):
    website = 'https://textbookcentre.com/catalogue/category/art-supplies/?page='+ str(i)
    
    
    #request
    response = requests.get(website)
    
    #soup
    soup = BeautifulSoup(response.content,'html.parser')
    
    #results
    results = soup.find_all('div', class_="product-item d-flex col-6 col-sm-4 col-lg-3")
    
    
    #loop
    
    for result in results:
        art_name.append(result.find('h5', {'class': 'card-title product-card-name'}).get_text())
        
        
    for result in results:
        price.append(result.find('span',{'class' : 'stockrecord-price-current'}).get_text().strip())
        
        
df = pd.DataFrame({'Art Supplies':art_name,'Art Price':price})
        
        
        
    
    
    
```
